```python
import pandas as pd
import numpy as np
```

```python
User=pd.read_csv("User.csv")
Content=pd.read_csv("Content.csv")
Session=pd.read_csv("Session.csv")
ReactionT=pd.read_csv("ReactionTypes.csv")
Profile=pd.read_csv("Profile.csv")
Location=pd.read_csv("Location.csv")
Reaction=pd.read_csv("Reactions.csv")
```

```python
User.drop('Unnamed: 0',axis=1,inplace=True)
Session.drop(['Unnamed: 0'],axis=1,inplace=True)

Profile.drop('Unnamed: 0',axis=1,inplace=True)
Location.drop('Unnamed: 0',axis=1,inplace=True)
ReactionT.drop('Unnamed: 0',axis=1,inplace=True)
Content.drop(['URL'],axis=1,inplace=True)
```

```python
# split column and add new columns to df
Location[['Street', 'Zip Code']] = Location['Address'].str.split(',', expand=True)
# display the dataframe
Location
```

```python
Profile['Interests'] =  Profile['Interests'].apply(lambda x: x.replace('[','').replace(']',''))
```

```python
Profile.head()
```

```python
data=Profile["Interests"].str.split(',', expand=True)
```

```python
data.rename(columns={0:"Interest 1", 1: "Interest 2", 2:"Interest 3", 3:"Interest 4"},inplace=True)
```

```python
data.head()
```

```python
Profile = pd.concat([data,Profile], axis=1)
 
```

```python
df = Profile.merge(Content,how='left').merge(Reaction,how ='left').merge(Location,how='left').merge(Session,how='left')
```

```python
df = Content.merge(Profile, how='left', on='User ID').merge(Location)
df = df.merge(Session, how='left', on='User ID')
df = df.merge(Reaction, how='left', on='Content ID')
df = df.rename(columns={'Type_x':'Cont_Type','Type_y':'Type'})
df = df.merge(ReactionT, how='left', on='Type')
df = df.rename(columns={'Type':'React_Type'})
df = df
```
